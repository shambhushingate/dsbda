```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Placement_Prediction_data.csv')
df
```

```python
df=df.drop(columns=["Unnamed: 0"])
df
```

```python
sns.countplot(data=df, x="PlacementStatus")
plt.title('Placement Status Count')
plt.show()
```

```python
sns.boxplot(data=df, x='PlacementStatus', y='CGPA', )
plt.title('CGPA vs Placement Status')
plt.show()
```

```python
sns.scatterplot(data=df, x='CGPA', y='Communication Skill Rating', hue='PlacementStatus')
plt.title('CGPA vs Communication Skill Rating')
plt.show()
```

```python
sns.countplot(data=df, x='Internship', hue='PlacementStatus')
plt.title('Number of Internships vs Placement')
plt.show()
```

```python
df_hackathon = pd.crosstab(df['Hackathon'], df['PlacementStatus'])
df_hackathon.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Hackathon Participation vs Placement')
plt.xlabel('Hackathon Participation')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.hist(df['12th Percentage'], bins=10, alpha=0.5, label='12th Percentage')

plt.legend()
plt.title('12th and 10th Percentages Distribution')
plt.show()
```

```python
corr = df[['CGPA', 'Major Projects', 'Mini Projects', 'Skills', 'Communication Skill Rating', 
           '12th Percentage', '10th Percentage']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1)
```

```python
cert_counts = df['Major Projects'].value_counts()
cert_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('viridis', len(cert_counts)), startangle=90)
plt.title('Workshop/Certification Count')
plt.ylabel('')
plt.show()
```

```python
# Pie Chart â€“ Placement Status Distribution
plt.figure(figsize=(6, 6))
placement_status_counts = df['PlacementStatus'].value_counts()
placement_status_counts.plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette('Set2', len(placement_status_counts)), startangle=90)
plt.title('Placement Status Distribution')
plt.ylabel('')
plt.show()
```

```python
# Violin Plot â€“ Communication Skill Rating vs Placement Status
plt.figure(figsize=(6, 4))
sns.violinplot(data=df, x='PlacementStatus', y='Communication Skill Rating', palette='muted')
plt.title('Communication Skill Rating vs Placement Status')
plt.show()
```

```python
# Bubble Plot â€“ Communication Skill vs CGPA (Bubble Size by Backlogs)
plt.figure(figsize=(8, 6))
plt.scatter(df['CGPA'], df['Communication Skill Rating'], alpha=0.5, c=df['PlacementStatus'].map({'Placed': 'green', 'NotPlaced': 'red'}))
plt.title('Communication Skill vs CGPA (Bubble Size by Backlogs)')
plt.xlabel('CGPA')
plt.ylabel('Communication Skill Rating')
plt.show()
```
